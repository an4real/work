```python
import os
import json
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
from tqdm import tqdm as tq  # Используем обычный tqdm
import time
```

```python
#старый вариант парсера

# качаем и делим на три таблички: с дпт, с авто и с людьми
for link in tq(json_links):
    r = requests.get(link)
    with open(link.split('/')[-1], 'w') as f:
        json.dump(r.json(), f)
    region_accs = r.json()

    accidents = []
    vehicles= []
    participants = []

    for acc in region_accs['features']:
        acc_pr = acc['properties']
        accident = {'id': acc_pr['id'],
                    'tags': ', '.join(acc_pr['tags']),
                    'category': acc_pr['category'],
                    'region': acc_pr['parent_region'],
                    'county': acc_pr['region'],
                    'address': acc_pr['address'],
                    'longitude': acc['geometry']['coordinates'][0],
                    'latitude': acc['geometry']['coordinates'][1],
                    'nearby': ', '.join(acc_pr['nearby']),
                    'datetime': acc_pr['datetime'],
                    'light': acc_pr['light'],
                    'weather': ', '.join(acc_pr['weather']),
                    'road_conditions': ', '.join(acc_pr['road_conditions']),
                    'participants_count': acc_pr['participants_count'],
                    'participant_categories': ', '.join(acc_pr['participant_categories']),
                    'severity': acc_pr['severity'],
                    'dead_count': acc_pr['dead_count'],
                    'injured_count': acc_pr['injured_count']}
        accidents.append(accident)

        for index_v, veh in enumerate(acc_pr['vehicles']):
            vehicle = {'accident_id': acc_pr['id'],
                       'vehicle_id': f"{acc_pr['id']}_{index_v+1}",
                       'category': veh['category'],
                       'brand': veh['brand'],
                       'model': veh['model'],
                       'color': veh['color'],
                       'year': veh['year'],}
            vehicles.append(vehicle)
            for index_p, person in enumerate(veh['participants']):
                participant = {'accident_id': acc_pr['id'],
                               'vehicle_id': f"{acc_pr['id']}_{index_v+1}",
                               'participant_id': f"{acc_pr['id']}_{index_v+1}_{index_p+1}",
                               'role': person['role'],
                               'gender': person['gender'],
                               'violations': ', '.join(person['violations']),
                               'health_status': person['health_status'],
                               'years_of_driving_experience': person['years_of_driving_experience']}
                participants.append(participant)
        for index_p, person in enumerate(acc_pr['participants']):
            participant = {'accident_id': acc_pr['id'],
                           'vehicle_id': None,
                           'participant_id': f"{acc_pr['id']}_0_{index_p+1}",
                           'role': person['role'],
                           'gender': person['gender'],
                           'violations': ', '.join(person['violations']),
                           'health_status': person['health_status'],
                           'years_of_driving_experience': None}
            participants.append(participant)
    pd.DataFrame(accidents).to_csv(link.split('/')[-1].split('.')[0]+'_accidents.csv', sep=';')
    pd.DataFrame(vehicles).to_csv(link.split('/')[-1].split('.')[0]+'_vehicles.csv', sep=';')
    pd.DataFrame(participants).to_csv(link.split('/')[-1].split('.')[0]+'_participants.csv', sep=';')
    
    time.sleep(2)
```

```python
#обновленный вариант парсера

import requests
import zipfile
import io
import json
import pandas as pd
import time
from tqdm import tqdm as tq

max_retries = 3  # максимальное количество попыток
timeout_sec = 60  # увеличенный тайм-аут

for link in tq(json_links):
    success = False
    for attempt in range(1, max_retries+1):
        try:
            r = requests.get(link, timeout=timeout_sec)
            
            if r.status_code != 200:
                print(f"Ошибка {r.status_code} при запросе {link}")
                break  

            if not link.endswith(".zip"):
                print(f"Файл {link} не является ZIP")
                break

            try:
                z = zipfile.ZipFile(io.BytesIO(r.content))
            except zipfile.BadZipFile:
                print(f"Файл {link} не удалось открыть как ZIP")
                break

            geojson_name = z.namelist()[0]
            with z.open(geojson_name) as f:
                region_accs = json.load(f)

            with open(geojson_name, 'w', encoding='utf-8') as f:
                json.dump(region_accs, f, ensure_ascii=False)

            accidents, vehicles, participants = [], [], []

            for acc in region_accs.get('features', []):
                acc_pr = acc['properties']

                
                accident = {
                    'id': acc_pr['id'],
                    'tags': ', '.join(acc_pr.get('tags', [])),
                    'category': acc_pr.get('category'),
                    'region': acc_pr.get('parent_region'),
                    'county': acc_pr.get('region'),
                    'address': acc_pr.get('address'),
                    'longitude': acc['geometry']['coordinates'][0],
                    'latitude': acc['geometry']['coordinates'][1],
                    'nearby': ', '.join(acc_pr.get('nearby', [])),
                    'datetime': acc_pr.get('datetime'),
                    'light': acc_pr.get('light'),
                    'weather': ', '.join(acc_pr.get('weather', [])),
                    'road_conditions': ', '.join(acc_pr.get('road_conditions', [])),
                    'participants_count': acc_pr.get('participants_count'),
                    'participant_categories': ', '.join(acc_pr.get('participant_categories', [])),
                    'severity': acc_pr.get('severity'),
                    'dead_count': acc_pr.get('dead_count'),
                    'injured_count': acc_pr.get('injured_count')
                }
                accidents.append(accident)

             
                for index_v, veh in enumerate(acc_pr.get('vehicles', [])):
                    vehicle_id = f"{acc_pr['id']}_{index_v+1}"
                    vehicle = {
                        'accident_id': acc_pr['id'],
                        'vehicle_id': vehicle_id,
                        'category': veh.get('category'),
                        'brand': veh.get('brand'),
                        'model': veh.get('model'),
                        'color': veh.get('color'),
                        'year': veh.get('year'),
                    }
                    vehicles.append(vehicle)

                    
                    for index_p, person in enumerate(veh.get('participants', [])):
                        participant = {
                            'accident_id': acc_pr['id'],
                            'vehicle_id': vehicle_id,
                            'participant_id': f"{acc_pr['id']}_{index_v+1}_{index_p+1}",
                            'role': person.get('role'),
                            'gender': person.get('gender'),
                            'violations': ', '.join(person.get('violations', [])),
                            'health_status': person.get('health_status'),
                            'years_of_driving_experience': person.get('years_of_driving_experience')
                        }
                        participants.append(participant)

                
                for index_p, person in enumerate(acc_pr.get('participants', [])):
                    participant = {
                        'accident_id': acc_pr['id'],
                        'vehicle_id': None,
                        'participant_id': f"{acc_pr['id']}_0_{index_p+1}",
                        'role': person.get('role'),
                        'gender': person.get('gender'),
                        'violations': ', '.join(person.get('violations', [])),
                        'health_status': person.get('health_status'),
                        'years_of_driving_experience': None
                    }
                    participants.append(participant)

            #Сохраняем CSV
            base_name = geojson_name.split('.')[0]
            pd.DataFrame(accidents).to_csv(base_name+'_accidents.csv', sep=';', index=False)
            pd.DataFrame(vehicles).to_csv(base_name+'_vehicles.csv', sep=';', index=False)
            pd.DataFrame(participants).to_csv(base_name+'_participants.csv', sep=';', index=False)

            time.sleep(2)
            success = True
            break  

        except requests.exceptions.ReadTimeout:
            print(f"⏱ Попытка {attempt} для {link} завершилась тайм-аутом, повтор...")
            time.sleep(5) 
            continue
        except Exception as e:
            print(f"Ошибка {link}: {e}")
            break

    if not success:
        print(f"Не удалось скачать {link} после {max_retries} попыток")
```

```python
import os
import pandas as pd

def concat_csv_files(pattern, output_file):
    df = pd.DataFrame()
    for file in [f for f in os.listdir() if pattern in f]:
        tmp = pd.read_csv(file, sep=';')
      
        if 'Unnamed: 0' in tmp.columns:
            tmp = tmp.drop(columns=['Unnamed: 0'])
        df = pd.concat([df, tmp], ignore_index=True)
    df.to_csv(output_file, index=False)
    return df

# Объединяем ДТП
accidents_df = concat_csv_files('_accidents.csv', 'accidents_all_regions.csv')

# Объединяем автомобили
vehicles_df = concat_csv_files('_vehicles.csv', 'vehicles_all_regions.csv')

# Объединяем участников
participants_df = concat_csv_files('_participants.csv', 'participants_all_regions.csv')
```
