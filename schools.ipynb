```python
import pandas as pd
import os


folder_path = r'C:\Users\1\Downloads\ФСН_№_ОО-2_за_2024_год\ОО-2 за 2024\(ГОУ) (город+село)'

# Список всех Excel-файлов в папке
files = [f for f in os.listdir(folder_path) if f.endswith('.xlsx')]

dfs = []

for file in files:
  
    file_path = os.path.join(folder_path, file)
    
    # Извлечь название региона до '(ГОУ)'
    region = file.split('(ГОУ)')[0].strip()
    
    try:
        # Прочитать лист 'Раздел 1.1' с заголовком из 17-й строки
        df = pd.read_excel(file_path, sheet_name='Раздел 1.1', header=17)
        
    
        df['Регион'] = region
        
      
        dfs.append(df)
        
        print(f"Файл обработан: {file}")
        
    except Exception as e:
        print(f"Ошибка с файлом {file}: {e}")

final_df = pd.concat(dfs, ignore_index=True)

final_df.to_excel('объединённые_данные.xlsx', index=False)

print("Все файлы объединены.")
```

```python
final_df = final_df.loc[:, ~final_df.columns.str.startswith('Unnamed')]
```

```python
import pandas as pd

final_df = final_df.dropna(subset=['Наименование\nпоказателей'])

final_df = final_df[~(
    (final_df['Наименование\nпоказателей'] == 1) |
    (final_df['Наименование\nпоказателей'].str.contains('Заполняется', na=False)) |
    (final_df['Наименование\nпоказателей'].str.contains('Кроме того', na=False)) |
    (final_df['Наименование\nпоказателей'].str.contains('Справка', na=False))
    
)]
final_df
```

```python
systems = [
    'Оборудовано водопроводом',
    'Обору-довано водоот-ведением (канали-зацией)',
    'Оборудовано центральным отоплением',
    'Доступно для маломо-бильных групп населения'
]

for system in systems:
    final_df[f'Не {system}'] = final_df['Всего'] - final_df[system]

final_df[['Регион', 'Всего', 'Требует капитального ремонта', 'Находится в аварийном состоянии' ] + [f'Не {s}' for s in systems]]
```

```python
cols_to_percent = [
    'Требует капитального ремонта',
    'Находится в аварийном состоянии',
    'Не Оборудовано водопроводом',
    'Не Обору-довано водоот-ведением (канали-зацией)',
    'Не Оборудовано центральным отоплением',
    'Не Доступно для маломо-бильных групп населения'
]

percent_df = final_df.copy()

for col in cols_to_percent:
    percent_df[col] = (percent_df[col] / percent_df['Всего']) * 100

percent_df[cols_to_percent] = percent_df[cols_to_percent].round(2)

percent_df[['Регион', 'Всего'] + cols_to_percent]
```

```python
percent_df.rename(columns={
    'Не Оборудовано водопроводом': 'Без водопровода',
    'Находится в аварийном состоянии': 'В аварийном состоянии',
    'Не Оборудовано центральным отоплением': 'Нет центрального отопления',
    'Не Доступно для маломо-бильных групп населения': 'Нет удобств для маломобильных',
    'Не Обору-довано водоот-ведением (канали-зацией)': 'Нет канализации',
    'Требует капитального ремонта': 'Нужен капремонт'
}, inplace=True)
```
